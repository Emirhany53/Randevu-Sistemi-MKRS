## Bernoulli Dağılımı

Bernoulli dağılımı, yalnızca **iki olasılıklı** (başarı veya başarısızlık) deneylerde kullanılır.

### Olasılık Kütle Fonksiyonu (PMF)

\[
P(X = x) = 
\begin{cases}
p & \text{eğer } x = 1 \ (\text{başarı}) \\
1 - p & \text{eğer } x = 0 \ (\text{başarısızlık})
\end{cases}
\]

Kısa hali:

\[
P(X = x) = p^x (1 - p)^{1 - x}, \quad x \in \{0, 1\}
\]

---

### Beklenen Değer (Ortalama)

\[
E(X) = p
\]

### Varyans

\[
Var(X) = p(1 - p)
\]

---

**Parametreler:**
- \( p \): Başarı olasılığı \((0 \leq p \leq 1)\)
- \( X \): 0 (başarısızlık) veya 1 (başarı) alabilen rassal değişken


```python
from scipy.stats import bernoulli
```

```python
p=0.6
result=bernoulli(p)           #Dağılımı burada tanımladık
result.pmf(k=0)               #pmf=probability mass function (olasılık kütle fonksyonu) 
#Bernoulli dağılımına göre, 0 (başarısızlık) sonucunun olasılığı nedir? sorusu cevaplanıyor
```

```python
Bir rassal değişkenin uzun vadeli kararlığını tanımlayan olasılık teoremidir.
```

```python
import numpy as np
rng=np.random.RandomState(123) #Rastgele sayı üreticisi (seed) sabitleniyor. Böylece her çalıştırmada aynı rastgele sonuçlar elde edilir.
for i in np.arange(1,21):
    deney_sayisi=2**i
    yazi_turalar=rng.randint(0,2,size=deney_sayisi)     #0 ve 1’den rastgele sayılar üretir. 0 → Tura, 1 → Yazı olarak kabul ediliyor.  
    yazi_olasılıkları=np.mean(yazi_turalar)
    print("Atış Sayısı:",deney_sayisi,"---","Yazı Olasılığı %.2f" % (yazi_olasılıkları*100))
```

### Binom Dağılımı

**Tanım:**

Binom dağılımı, sabit sayıda bağımsız denemede, her bir denemede aynı başarı olasılığı ile belirli sayıda başarının gerçekleşme olasılığını verir.

Bir deneme:
- İki olasılıklıdır: **başarı (1)** veya **başarısızlık (0)**  
- Denemeler **birbirinden bağımsızdır**  
- Her denemede başarı olasılığı **aynıdır (p)**  

---

#### Olasılık Kütle Fonksiyonu (PMF):

$$
P(X = k) = \binom{n}{k} \cdot p^k \cdot (1 - p)^{n - k}
$$

Burada:
- \( n \): Deneme sayısı  
- \( k \): Başarı sayısı  
- \( p \): Başarı olasılığı  
- \( \binom{n}{k} \): "n'in k'li kombinasyonu" yani \( \frac{n!}{k!(n - k)!} \)

---

#### Beklenen Değer (Ortalama):

$$
E(X) = n \cdot p
$$

#### Varyans:

$$
Var(X) = n \cdot p \cdot (1 - p)
$$


```python
from scipy.stats import binom
```

```python
p=0.01
n=100
rv=binom(n,p)
print(rv.pmf(1)) #100 denemede 1 kez başarılı olma olasılığı
print(rv.pmf(10)) #100 denemede 10 kez başarılı olma olasılığı
```

### Poisson Dağılımı

Poisson dağılımının olasılık kütle fonksiyonu (PMF) şu şekildedir:

$$
P(X = k) = \frac{e^{-\lambda} \cdot \lambda^k}{k!}
$$

Burada:
- \( \lambda \): Belirli bir zaman aralığında beklenen olay sayısı (ortalama)
- \( k \): Olay sayısı (0, 1, 2, ...)
- \( e \): Euler sabiti (yaklaşık 2.718)

---

#### Beklenen Değer (Ortalama):

$$
E(X) = \lambda
$$

#### Varyans:

$$
Var(X) = \lambda
$$


```python
from scipy.stats import poisson

λ=0.1 
rv=poisson(mu=λ)       #ortalama olay sayısı 0.1 olan bir dağılım oluşturuldu.
print(rv.pmf(k=0))
print(rv.pmf(k=5))     #Ortalama olay sayısı 0.1 olan dağılımda olay sayısısının 5 olma olasılığı
```

$$
f(x) = \frac{1}{\sqrt{2\pi\sigma^2}} \cdot e^{-\frac{(x - \mu)^2}{2\sigma^2}}
$$

### Z-Testi Formülü

$$
Z = \frac{\bar{x} - \mu}{\frac{\sigma}{\sqrt{n}}}
$$

- \( \bar{x} \): Örneklem ortalaması  
- \( \mu \): Anakütle ortalaması (hipotezdeki değer)  
- \( \sigma \): Anakütle standart sapması  
- \( n \): Örneklem büyüklüğü
### T-Testi Formülü

$$
t = \frac{\bar{x} - \mu}{\frac{s}{\sqrt{n}}}
$$

- \( \bar{x} \): Örneklem ortalaması  
- \( \mu \): Anakütle ortalaması (hipotezdeki değer)  
- \( s \): Örneklem standart sapması  
- \( n \): Örneklem büyüklüğü

# İŞ Uygulaması: Dönüşüm Oranı Testi  

**Problem:** Bir yazılım ile bir mecrada reklam verilmiş ve bu reklama ilişkin yazılım tarafından 0.125 dönüşüm oranı elde edildiği ifade edilmiş.  
Fakat bu durum kontrol edilmek isteniyor. Çünkü bu yüksek bir oran ve gelirler incelendiğinde örtüşmüyor.  

**Detaylar:** 500 kişi dış mecrada reklamlara tıklamış, 40 tanesi sitemize gelip alışveriş yapmış.  

**Örnek üzerinden elde edilen dönüşüm oranı:**  
\[
\frac{40}{500} = 0.08
\]

**Hipotezler:**  
- \(H_0: p = 0.125\) (gerçek dönüşüm oranı %12.5’tir)  
- \(H_1: p \neq 0.125\) (gerçek dönüşüm oranı %12.5 değildir)  


```python
from statsmodels.stats.proportion import proportions_ztest
```

```python
count=40     #Başarılı gözlem
nobs=500     #Gözlem Sayısı
value=0.125  #Test etmek istediğimiz değer
```

```python
proportions_ztest(count,nobs,value)
```
